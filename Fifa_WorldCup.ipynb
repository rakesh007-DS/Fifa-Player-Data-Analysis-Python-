```python
import pandas as pd
eda=pd.read_csv("eda.csv",encoding="latin-1")
```

```python
eda.tail(5)
```

```python

print("Shape:", eda.shape)
print("Number of Duplicates:", eda.duplicated().sum())
print("\nColumns:", eda.columns.tolist())
print("\nData Types:\n", eda.dtypes)
print("\nMissing Values:\n", eda.isnull().sum())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(eda['Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Count of Players")
plt.title("Distribution of Players' Age")
plt.show()
```

```python
import seaborn as sns

cols = ['Stamina', 'Strength', 'Acceleration','SprintSpeed',
        'Agility', 'Jumping' ,'Vision','Reactions']

plt.figure(figsize=(15,12))
for i, col in enumerate(cols, 1):
    plt.subplot(3,3,i)
    sns.regplot(data=eda, x="Age", y=col, scatter_kws={'s':10}, line_kws={'color':'red'})
    plt.title(f"Age vs {col}")
plt.tight_layout()
plt.show()
```

```python
sorted_players = eda.sort_values(by="Age", ascending=False)[['Name','Age','Nationality','Club']]
print(sorted_players.head(20))   # display top 20 oldest
```

```python
popular_clubs = ['Juventus','Liverpool','Ajax','Manchester City','Paris Saint-Germain',
                 'FC Bayern München','Real Madrid','FC Barcelona','Manchester United','FC Porto']

plt.figure(figsize=(12,6))
sns.violinplot(data=eda[eda['Club'].isin(popular_clubs)], x="Club", y="Age")
plt.xticks(rotation=45)
plt.title("Age Distribution in Popular Clubs")
plt.show()
```

```python
top20 = eda.sort_values(by="Overall", ascending=False).head(20)

plt.figure(figsize=(12,6))
sns.barplot(data=top20, x="Overall", y="Name", palette="viridis")
plt.title("Top 20 Players by Overall Rating")
plt.xlabel("Overall Rating")
plt.ylabel("Player Name")
plt.show()

```
